- Seaborn is an open source, BSD-licensed Python library providing high level API for visualizing the data using Python programming language.

- It is built on top of Matplotlib.

#### SeabornVs Matplotlib
It is summarized that if Matplotlib “tries to make easy things easy and hard things possible”, Seaborn tries to make a well-defined set of hard things easy too.”

Seaborn helps resolve the two major problems faced by Matplotlib; the problems are −

- 1.Default Matplotlib parameters
- 2.Working with data frames

##### Histogram using distance plotting

Histograms represent the data distribution by forming bins along the range of the data and then drawing bars to show the number of observations that fall in each bin.

# Univariate bivariate multivariate???

#### Plotting Biviariate Distribution:Joint Plot

Bivariate Distribution used to determine the relation between two variables.

```python
# It gives scatter plot and its strictly bivariate plot
sns.jointplot(x='sepal_length',y='sepal_width',data=i)
```

```python
# use kind= hex
sns.jointplot(x='sepal_length',y='sepal_width',data=i,kind='hex')
```

```python
# Assignment: Use regression plot kind='reg'
sns.jointplot(x='sepal_length',y='sepal_width',data=i,kind='reg')
```

```python
# categoriwise scatter plot: make it multivariate
# using an attribute hue
sns.jointplot(x='sepal_length',y='sepal_width',data=i,kind='scatter',hue='species')
```

```python
# change colors
sns.jointplot(x='sepal_length',y='sepal_width',data=i,kind='scatter',hue='species',palette='afmhot')
```

##### Kernel Density Estimation

Kernal density estimation is non-parametic way to estimate the distribution of a variable.

```python
import seaborn as sns
```

```python
# kde is univaraite plot
# works only on numerics data
#load iris dataset
i = sns.load_dataset('iris')
i.head()
```

```python
# let's plot kde for all numeric columns 
sns.kdeplot(data=i)
```

```python
sns.kdeplot(x='sepal_length',data=i)
```

##### Pair relationships using pairplot

To plot multiple pairwise bivariate distribution in a dataset,you can use the pairplot() function.

```python
#Pairplot works on Numeric Data only.
sns.pairplot(data=i)
```

```python
#hue is used on categorical data
sns.pairplot(data=i,hue='species')
```

```python
sns.pairplot(data=i[['sepal_length','petal_length','species']],hue='species')
```

```python
sns.pairplot(data=i[['sepal_length','petal_length','species']],hue='species',kind='kde')
```
